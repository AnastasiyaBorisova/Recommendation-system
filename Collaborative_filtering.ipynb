```python
def jaccard(x1, x2):
  x1 = df[x1].to_numpy()
  x2 = df[x2].to_numpy()
  o = 0
  for i in range(len(x1)):
    if x1[i] == x2[i]:
      o += 1
  return o / len(x1)
print(jaccard('–®–∞–º–ø—É–Ω—å', '–ö–æ–Ω–¥–∏—Ü–∏–æ–Ω–µ—Ä'))
```
