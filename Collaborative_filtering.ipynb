Расстояние Жаккара

```python
def jaccard(x1, x2):
  x1 = df[x1].to_numpy()
  x2 = df[x2].to_numpy()
  o = 0
  for i in range(len(x1)):
    if x1[i] == x2[i]:
      o += 1
  return o / len(x1)
print(jaccard('Шампунь', 'Кондиционер'))
```

Измерение расстояния с помощью 𝐿𝑝-норм (Манхэттенское)

```python
def L1(x1, x2):
  x1 = df[x1].to_numpy()
  x2 = df[x2].to_numpy()
  return np.sum(abs(x1-x2))
print(L1('Шампунь', 'Кондиционер'))
```

Измерение расстояния с помощью 𝐿𝑝-норм (Евклидово)

```python
import math
def L2(x1, x2):
  x1 = df[x1].to_numpy()
  x2 = df[x2].to_numpy()
  return math.sqrt(np.sum(abs(x1-x2)**2))
print(L2('Шампунь', 'Кондиционер'))
```

Коэффициент Отиаи

```python
def otiai(x1, x2):
  x1 = df[x1].to_numpy()
  x2 = df[x2].to_numpy()
  x1x2 = 0
  x_1 = 0
  x_2 = 0
  for i in range(len(x1)):
     x1x2 += x1[i] * x2[i]
     x_1 += x1[i]**2
     x_2 += x2[i]**2
  return x1x2/(math.sqrt(x_1) * math.sqrt(x_2))
print(otiai('Шампунь', 'Кондиционер'))
```
